```python
import gff3_parser
filepath = r"../../data/annotations/Sparus_aurata.fSpaAur1.1.113.gff3"
sparus_biomart_annotations = gff3_parser.parse_gff3(filepath, verbose = True, parse_attributes = True)
```

```python
merged_data_path = r"../../data/plots/filtered_merged_data.csv"
```

```python
import pandas as pd
data = pd.read_csv(merged_data_path)
```

```python
data
```

```python
chr_positions = list(zip(data['CHR'], data['POS']))
for chr_pos in chr_positions:
    print(chr_pos[0], chr_pos[1])
```

```python
# keep only the rows for genes
sparus_biomart_annotations = sparus_biomart_annotations[(sparus_biomart_annotations["Type"]=="gene") or (sparus_biomart_annotations["Type"]=="biological_region" and )]
sparus_biomart_annotations = sparus_biomart_annotations[(sparus_biomart_annotations["Type"]=="gene")]
chr_positions = list(zip(data['CHR'], data['POS']))
a = []
# 1-25 : LR537121 - LR537144
for LG_ID in range(1, 25):
    for CHR, SNP_position in chr_positions:
        # CHR = chr_pos[0]
        # SNP_position = chr_pos[1]
        print (CHR, SNP_position, LG_ID)
        if CHR == LG_ID+20:
            sparus_biomart_annotations1 = sparus_biomart_annotations[(sparus_biomart_annotations["Seqid"]==str(CHR-20))]
            sparus_biomart_annotations1["Start"] = sparus_biomart_annotations["Start"].astype('int')
            sparus_biomart_annotations1["End"] = sparus_biomart_annotations["End"].astype('int')
            print("SNP_position: ", SNP_position)
            print("LG_ID: ", LG_ID)
            a = sparus_biomart_annotations1[(sparus_biomart_annotations1["Start"] > SNP_position - 200000) &
                                            (sparus_biomart_annotations1["End"] < SNP_position + 200000)]
            # print(a)
            a.to_csv(fr"../../data/annotations/Filtered LR5371{CHR} Pos {SNP_position} gene annotations Sparus aurata poolseq.csv", index=False)
```
